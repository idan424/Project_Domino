```python
import os
import tqdm

path = r'/content/drive/MyDrive/Colab Notebooks/Project Domino/ch_names extract/'

file_names = [name for name in os.listdir(path) if 'vhdr' in name]
```

```python
def get_data(file_name):
    # print(file_name)
    subj_name = file_name[:7]
    elec_type = file_name[-10:-5]
    with open(path + file_name) as f:
        content = f.readlines()

    ch_names = [line.strip(', \n').split('=') for line in content if 'Ch' == line[0:2]]
    return subj_name, ch_names, elec_type

def write_data(subj_name, ch_names, elec_type):
    with open(path + 'Electrode Names.txt', 'a') as f:
        
        for ch_n in ch_names:
            f.write(f'{subj_name},{ch_n[0]},{ch_n[1]},{elec_type}\n')
```

```python
for file_name in tqdm.tqdm(file_names): 
    write_data(*get_data(file_name))
```
